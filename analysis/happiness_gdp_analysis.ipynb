De relatie tussen economische welvaart, geluk en levensverwachting is een veelbesproken onderwerp binnen de economie, sociologie en volksgezondheid. Volgens de economische groei-theorie van Richard Easterlin leidt een hoger bruto binnenlands product (BBP) per hoofd van de bevolking in armere landen tot een stijging in welzijn, omdat basisbehoeften zoals voedsel, onderdak en gezondheidszorg beter worden vervuld. Echter, in welvarende landen neemt de invloed van BBP op geluk af, aangezien materiële vooruitgang niet altijd leidt tot meer subjectief welzijn. Dit fenomeen staat bekend als de Easterlin Paradox (Easterlin, 1974).

Daarnaast stelt de capability approach van Amartya Sen (1999) dat economische groei pas leidt tot welzijn als deze zich vertaalt in verbeterde gezondheidszorg, onderwijs en sociale zekerheid. Dit sluit aan bij studies van Bloom & Canning (2000), die aantonen dat een hoger BBP per hoofd van de bevolking samenhangt met een langere levensverwachting, voornamelijk door investeringen in gezondheidsvoorzieningen en infrastructuur.

De gegevens uit het World Happiness Report bevestigen deze theorieën en laten een sterke correlatie zien tussen economische prestaties en welzijnsindicatoren. Toch benadrukken onderzoekers zoals Helliwell et al. (2021) dat sociale steun, vrijheid en vertrouwen in de overheid minstens even belangrijk zijn voor geluk en gezondheid. Dit onderzoek richt zich op het kwantificeren van deze relaties en het bepalen in hoeverre BBP per hoofd van de bevolking een doorslaggevende factor is in geluk en levensverwachting, in vergelijking met andere sociale en economische variabelen.

#### Omgaan met ontbrekende data

Bij het combineren van datasets ontstaan vaak ontbrekende waarden, omdat de ene dataset niet altijd dezelfde kolommen en rijen bevat als de andere. Dit probleem is een veelvoorkomend fenomeen in data-analyse en vormt een uitdaging voor onderzoekers, datawetenschappers en analisten die streven naar complete en bruikbare datasets. Ontbrekende waarden kunnen optreden door verschillende oorzaken, zoals fouten in de dataverzameling, inconsistenties tussen verschillende databronnen of het ontbreken van relevante gegevens in een bepaalde context. Het omgaan met deze ontbrekende waarden is een cruciale stap in het data-analyseproces, aangezien de keuzes die worden gemaakt direct van invloed zijn op de kwaliteit, nauwkeurigheid en interpretatie van de resultaten.

Een veelgebruikte oplossing voor ontbrekende waarden is het verwijderen ervan. Dit kan op het eerste gezicht een eenvoudige en snelle methode lijken, maar het heeft aanzienlijke nadelen. Het verwijderen van rijen of kolommen met ontbrekende waarden kan leiden tot een verlies van waardevolle informatie, vooral als de ontbrekende gegevens een groot deel van de dataset omvatten. Dit kan de representativiteit van de data beïnvloeden, waardoor de uiteindelijke analyses minder betrouwbaar worden. Bovendien kan deze aanpak leiden tot bias, met name wanneer de ontbrekende waarden niet willekeurig zijn, maar gerelateerd zijn aan een specifieke eigenschap van de data. Bijvoorbeeld, als in een dataset over inkomen en geluk de gegevens van mensen met lagere inkomens vaker ontbreken, kan het verwijderen van deze rijen een vertekend beeld geven van de relatie tussen inkomen en geluk.

Een alternatieve aanpak is het toepassen van imputatiemethoden, waarbij ontbrekende waarden worden ingevuld op basis van schattingen of voorspellingen. Er zijn verschillende technieken beschikbaar, variërend van eenvoudige methoden zoals het invullen met het gemiddelde of de mediaan, tot geavanceerdere methoden zoals regressie-imputatie of machine learning-algoritmen. De keuze van de methode hangt sterk af van de aard van de dataset, de variabelen en het doel van de analyse.

Het invullen met het gemiddelde of de mediaan is een van de meest gebruikte methoden vanwege de eenvoud. Deze aanpak is echter niet altijd geschikt, vooral niet als de data sterke trends of verdelingen vertoont. Een voorbeeld hiervan is het invullen van ontbrekende temperatuurwaarden: als de temperatuurgegevens seizoensgebonden fluctueren, kan het gebruik van een algemeen gemiddelde de data vertekenen. In dergelijke gevallen is het beter om de imputatie aan te passen aan de specifieke patronen in de data, bijvoorbeeld door historische trends of seizoensgebonden gemiddelden te gebruiken.

Een andere veelgebruikte methode is regressie-imputatie, waarbij de ontbrekende waarden worden voorspeld op basis van andere variabelen in de dataset. Deze methode maakt gebruik van correlaties en patronen tussen variabelen om realistische schattingen te genereren. Stel bijvoorbeeld dat er in een dataset een sterke correlatie bestaat tussen GDP per capita en de Happiness Score. In dit geval kan regressie worden gebruikt om de ontbrekende waarden in de Happiness Score te schatten op basis van de beschikbare GDP-gegevens. Hoewel deze methode vaak nauwkeuriger is dan eenvoudige imputatietechnieken, brengt het ook risico's met zich mee. Zo kunnen de voorspellingen de bestaande correlaties versterken of verzwakken, afhankelijk van hoe de regressie is toegepast en geëvalueerd.

Meer geavanceerde technieken maken gebruik van machine learning-algoritmen, zoals Random Forest of k-Nearest Neighbors (k-NN), om ontbrekende waarden in te vullen. Deze methoden zijn bijzonder krachtig bij datasets met complexe interacties tussen variabelen. Machine learning-modellen kunnen patronen in de data identificeren die voor mensen moeilijk waarneembaar zijn en deze gebruiken om nauwkeurige schattingen te genereren. Het nadeel van deze methoden is echter dat ze vaak computationeel intensief zijn en specifieke expertise vereisen om correct te worden toegepast.

In praktische toepassingen, zoals het invullen van ontbrekende waarden in variabelen zoals de Happiness Score, kunnen correlaties met andere variabelen een belangrijke rol spelen. Variabelen zoals GDP per capita, sociale steun en vrijheid worden vaak gebruikt als indicatoren om ontbrekende waarden te schatten. Het is hierbij essentieel om te begrijpen welke variabelen de meeste invloed hebben op de ontbrekende waarden en hoe deze variabelen zich onderling verhouden. Dit vraagt om een grondige verkenning van de data en, indien nodig, overleg met experts of docenten die bekend zijn met het vakgebied.

Het omgaan met ontbrekende waarden in datasets is een complex maar essentieel proces dat zowel technische vaardigheden als een goed begrip van de context vereist. Hoewel er geen universele oplossing is, biedt een combinatie van statistische en machine learning-methoden een breed scala aan mogelijkheden om dit probleem aan te pakken. Het belangrijkste is om een evenwicht te vinden tussen nauwkeurigheid, bruikbaarheid en transparantie, zodat de data een betrouwbare basis biedt voor verdere analyse en besluitvorming.

**De voorbereiding van de dataset bestond uit de volgende stappen:**

2. Elke dataframe kreeg een nieuwe kolom die het bijbehorende jaar aangeeft.

```python
df_2015['Year'] = 2015
df_2016['Year'] = 2016
df_2017['Year'] = 2017
df_2018['Year'] = 2018
df_2019['Year'] = 2019

combined_df = pd.concat([df_2015, df_2016, df_2017, df_2018, df_2019], ignore_index=True)
```

```python
columns = ['Year'] + [col for col in combined_df.columns if col != 'Year']
combined_df = combined_df[columns]
```

```python
combined_df['Economy_GDP_per_Capita'] = combined_df['Economy (GDP per Capita)'].fillna(combined_df['Economy..GDP.per.Capita.'])

combined_df.drop(columns=['Economy (GDP per Capita)', 'Economy..GDP.per.Capita.'], inplace=True)

combined_df['Trust_Government_Corruption'] = combined_df['Trust (Government Corruption)'].fillna(combined_df['Trust..Government.Corruption.'])

combined_df.drop(columns=['Trust (Government Corruption)', 'Trust..Government.Corruption.'], inplace=True)

combined_df['Trust_Government_Corruption'] = combined_df['Trust_Government_Corruption'].fillna(combined_df['Perceptions of corruption'])

combined_df['Country'] = combined_df['Country'].fillna(combined_df['Country or region'])
combined_df['Happiness Rank'] = combined_df['Happiness Rank'].fillna(combined_df['Overall rank'])
combined_df['Happiness Score'] = combined_df['Happiness Score'].fillna(combined_df['Score'])
combined_df['Economy_GDP_per_Capita'] = combined_df['Economy_GDP_per_Capita'].fillna(combined_df['GDP per capita'])
combined_df['Health_Life_Expectancy'] = combined_df['Health (Life Expectancy)'].fillna(combined_df['Healthy life expectancy'])

combined_df.drop(columns=['GDP per capita', 'Country or region', 'Score','Overall rank', 'Perceptions of corruption','Healthy life expectancy', 'Health (Life Expectancy)',], inplace=True)       

combined_df.drop(columns=['Region'], inplace=True)

combined_df['Freedom'] = combined_df['Freedom'].fillna(combined_df['Freedom to make life choices'])
combined_df['Social support'] = combined_df['Family'].fillna(combined_df['Social support'])

combined_df.drop(columns=['Freedom to make life choices', 'Family'], inplace=True)

combined_df['Happiness Rank'] = combined_df['Happiness Rank'].fillna(combined_df['Happiness.Rank'])
combined_df['Happiness Score'] = combined_df['Happiness Score'].fillna(combined_df['Happiness.Score'])
combined_df.drop(columns=['Happiness.Score'], inplace=True)
combined_df['Health_Life_Expectancy'] = combined_df['Health_Life_Expectancy'].fillna(combined_df['Health..Life.Expectancy.'])
combined_df.drop(columns=['Health..Life.Expectancy.'], inplace=True)
combined_df.drop(columns=['Dystopia Residual', 'Lower Confidence Interval', 'Upper Confidence Interval', 'Whisker.high', 'Whisker.low', 'Dystopia.Residual'], inplace=True)

combined_df.drop(columns=['Standard Error'], inplace=True)

combined_df.drop(columns=['Happiness.Rank'], inplace=True)
```

## Verdeling en Spreiding van de Data

```python
import numpy as np
important_vars = ['Happiness Score', 'Economy_GDP_per_Capita', 'Social support', 'Health_Life_Expectancy', 'Freedom']
subset = combined_df[important_vars]

analysis = {}
for var in important_vars:
    analysis[var] = {
        'mean': subset[var].mean(),
        'median': subset[var].median(),
        'mode': subset[var].mode().iloc[0],
        'std_dev': subset[var].std(),
        'variance': subset[var].var()
    }
for var in important_vars:
    plt.figure(figsize=(10, 5))
    sns.boxplot(x=subset[var])
    plt.title(f"Boxplot of {var}")
    plt.show()

    plt.figure(figsize=(10, 5))
    sns.histplot(subset[var], kde=True, bins=20)
    plt.title(f"Histogram of {var}")
    plt.show()
```

De histogrammen en boxplots geven een gedetailleerd beeld van de verdeling en spreiding van de belangrijkste variabelen in de dataset. Uit de histogrammen blijkt dat de Happiness Score een relatief normale verdeling heeft, maar met enkele landen die uitzonderlijk lage of hoge scores rapporteren. Dit suggereert dat een langere levensduur, vaak het gevolg van betere medische zorg en levensomstandigheden, een significante rol speelt in het welzijn van mensen, wat in lijn is met de bevindingen van Bloom & Canning (2000), die stellen dat investeringen in gezondheidszorg en infrastructuur bijdragen aan een hogere levensverwachting en welzijn.

Daarnaast laat de analyse een positieve correlatie zien tussen Happiness Score en Social Support (r = 0.65), wat aangeeft dat landen met sterkere sociale netwerken en meer onderlinge steun hogere geluksscores rapporteren. Deze bevinding sluit aan bij de capability approach van Amartya Sen (1999), die stelt dat welzijn meer omvat dan alleen economische factoren, en dat vrijheid essentieel is voor het realiseren van een goed leven.

Ten slotte blijkt uit de correlatieanalyse dat Happiness Score en Trust_Government_Corruption ook een positieve samenhang vertonen (r = 0.55). Samenvattend tonen de resultaten aan dat geluk niet door één enkele factor wordt bepaald, maar het resultaat is van een complex samenspel van economische, sociale, gezondheids- en bestuursaspecten, waarbij economische welvaart en levensverwachting de sterkste voorspellers lijken te zijn.

## Discussie  

Bij de analyse van de relatie tussen BBP per hoofd van de bevolking, geluk en levensverwachting kunnen verschillende vormen van bias de resultaten beïnvloeden. Een van de belangrijkste uitdagingen is culturele bias, waarbij de perceptie van geluk per samenleving kan verschillen. In sommige culturen wordt geluk primair als een individuele ervaring gezien, terwijl in andere het collectieve welzijn zwaarder weegt. Dit kan leiden tot vertekeningen in de gerapporteerde Happiness Scores, omdat bepaalde landen hogere scores kunnen rapporteren ondanks een lagere levensstandaard, simpelweg door een andere interpretatie van geluk.  

Daarnaast speelt economische bias een rol, aangezien BBP per hoofd van de bevolking als een centrale maatstaf wordt gebruikt om geluk te verklaren. Hoewel een hogere economische ontwikkeling over het algemeen gepaard gaat met een hoger welzijnsniveau, houdt deze indicator geen rekening met inkomensongelijkheid of subjectieve tevredenheid, die eveneens van invloed kunnen zijn op geluk. De dataset richt zich op nationale gemiddelden, wat kan maskeren hoe geluk binnen landen ongelijk verdeeld is.  

Een derde aandachtspunt is selectiebias, omdat de dataset mogelijk niet volledig representatief is voor alle landen. Kleine of economisch minder ontwikkelde landen kunnen ondervertegenwoordigd zijn, wat een vertekend beeld kan geven van de algemene correlaties. Bovendien bevat de dataset voornamelijk landen waarvoor gedetailleerde economische en sociale gegevens beschikbaar zijn, wat kan leiden tot een systematische vertekening in de interpretatie van de resultaten.  

Hoewel sommige van deze biases direct zichtbaar zijn in de dataset, kunnen andere slechts indirect worden vastgesteld via aanvullende analyses. Zo zou culturele bias verder onderzocht kunnen worden door subgroepen binnen de dataset te vergelijken op basis van regio of sociaaleconomische kenmerken. Dit benadrukt het belang van een kritische interpretatie van de resultaten en de noodzaak om mogelijke beperkingen in de dataset mee te wegen bij de conclusies.
